```python
import torch
import torch.nn as nn
import torch.nn.functional as F

import numpy as np
```

```python
# Read the test and train datasets
train_data = torch.load('data/train_dataset_64_all.pt')
test_data = torch.load('data/test_dataset_64_all.pt')
```

```python
d1 = list(train_data.data)
d2 = list(test_data.data)

l1 = list(train_data.labels)
l2 = list(test_data.labels)

# Concatenate the datasets
d1.extend(d2)
l1.extend(l2)

data = d1
labels = l1
```
